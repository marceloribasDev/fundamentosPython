# Introducao a Computacao na Nuvem

## Iass

Infrastructure as a service --> Iaas

- Computadores virtualizados, provem processamento, rede armazenamento
- Escalavel
- Normalmente nao necessita provisionamento

### O que voce pode contratar?

- Processamento
- Armazenamento
- Banco de dados
- Inteligencia artificial
- IOT



---



## PaaS

Plataform as a service --> PaaS

- Prove hardware e aplicativos
- Para desenvolvimento, prove ambiente com ferramentas e framewroks
- Multiplo ambientes e Sistemas Operacinais pelo mesmo custo
- Inclui servidores, armazenamento, rede, banco de dados etc...
- Atualizado



---



# SaaS

Software as a Service --> SaaS

- Provem aplicativos pela internet
- Empresa nao precisa instalar, hospedar ou manter software
- Prove acesso a tecnologia atualizada
- Forma mais comem de cloud computing
- Modelo de assinatura - por usuario/gratuito
- Prove todos os servicos relacionados
- Exemplos:
    - Office 365
    - Sales Force
    - Vyond

# Tipos de Nuvem

- Publico
    - Nuvem publica, contrata e utiliza
    - Oferecida na internet
    - Qualquer um pode contratar
    - Inclui hardware, rede, armazenamento, banco de dados, aplicacoes e etc...
    - Pago conforme o uso ou gratuito
    - Pode ter sua capaciade pernonalizado de acordo com a demanda
    - Exemplos:
        - AWS, Azure, Alibaba Cloud
- Privado
    - Nubem privada, sem acesso a internet
    - O cliente tem controle sobre toda a infraestrutura
    - Estruturada na infra da empres, nao e publica, nao esta na internet
    - Permite controle total sobre seus dados
    - VPC: Virtual Privite Cloud -> Eh quando um provedor dedica certa infra para um cliente especifico
    - VPC usa uma rede privada para prover servicos de nuvem na rede do cliente
    - O cliente pode estender sua rede usando a internet
    - Exemplos:
        - AWS, Azure e etc...

- Hibrido
    - mix entre publica e privada
